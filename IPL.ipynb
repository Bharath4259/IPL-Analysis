```python
matches = []
# print(mdf.head(2))
prev_index = 0
for year in range(2008, 2017):
    season = mdf.loc[mdf['season'] == year]
#     print(season)
    last_index = season.last_valid_index()   #.last_valid_index() will return last index
    last_row = mdf.iloc[[last_index]]                    #Data of last index
    teams = [year, last_row.loc[last_index].at['team1'], last_row.loc[last_index].at['team2'], last_row.loc[last_index].at['winner']]
    matches.append(teams)
    
# print(matches)
# Winners data of each IPL season
final_matches_df = pd.DataFrame(matches, columns=['year', 'team1', 'team2', 'winner_team'])
final_matches_df 
```

## How each team is performing in each city?

#### City wise winning data

```python
# Reading csv file again bcz we're getting a NameError showing " NameError: name 'mdf' is not defined "
# mdf = pd.read_csv("data/ipl_matches.csv")
city_mList = [mdf['winner'].tolist(), mdf['city'].tolist()]
x = map(list, zip(*city_mList))    # Getting a transpose 

city_matches = pd.DataFrame(list(x), columns=['winner', 'city'])
# city_matches

grouped_lst = city_matches[['winner', 'city']].groupby(['winner','city'], sort=True, group_keys=True)['city'].count()
grouped_lst = grouped_lst.reset_index(name ="count") #getting count of each cities they won
grouped_lst.head(50)
```

```python
# Teames with city they most won
max_win_list = grouped_lst.loc[grouped_lst.groupby(["winner"])['count'].idxmax()]
pd.DataFrame(max_win_list)  #To convert it into DataFrame
```

```python
fig, axes = plt.subplots(7,2,figsize=(50,150)) # Creating a subgraph with 7 X 2 size
j=0
k=-1
teamList = grouped_lst.winner.unique()
sns.set_context("paper",rc={'font.size': 32, 'axes.labelsize': 32, 'legend.fontsize': 32.0, 'axes.titlesize': 32, 'xtick.labelsize': 32, 'ytick.labelsize': 32})
for x in grouped_lst.winner.unique():
    data = grouped_lst.loc[grouped_lst['winner'] == x ]
    data2 =  data[['city', 'count']]
    k += 1
    if(k>1):
        k=0
    ax = sns.barplot(x='city', y='count', data = data2, color=team_colors[x], ax=axes[j, k])
#     ax.tick_params(labelsize=0.1)
    ax.set_title(x, fontsize=50)
    ax.set_xlabel('City')
    ax.set_ylabel('No of times won')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    if(k == 1):
        j += 1
        
```

## Individual Team performance when they won the toss

```python
mdf.head(10)[['team1', 'toss_winner', 'toss_decision', 'winner']].rename(columns={'team1':'team'})
```

```python
toss_data = pd.concat([ mdf[['team1', 'toss_winner', 'toss_decision', 'winner']].rename(columns={'team1':'team'}), 
          mdf[['team2', 'toss_winner', 'toss_decision', 'winner']].rename(columns={'team2':'team'})
          ], sort=True)
# toss_data
```

###### # Getting the winning performance list of each team when they won toss

```python
toss_performance = pd.DataFrame(toss_data.groupby(['winner', 'toss_winner'])['winner'].count().reset_index(name="count"))
toss_performance
# In below list count = the number of times the perticular team won the match
```

### Showing how winning of toss effecting the team performance with each team

```python
fig, axes = plt.subplots(7,2,figsize=(50,150))
j=0
k=-1
teamList = toss_performance.winner.unique()
sns.set_context("paper",rc={'font.size': 32, 'axes.labelsize': 32, 'legend.fontsize': 32.0, 'axes.titlesize': 32, 'xtick.labelsize': 32, 'ytick.labelsize': 32})
for x in teamList:
    data = toss_performance.loc[toss_performance['winner'] == x ]
    data2 =  data[['toss_winner', 'count']]
    k += 1
    if(k>1):
        k=0
    ax = sns.barplot(x='toss_winner', y='count', data = data2, color=team_colors[x], ax=axes[j, k])
#     ax.tick_params(labelsize=0.1)
    ax.set_title(x, fontsize=50)
    ax.set_xlabel('toss_winner')
    ax.set_ylabel('No of times won')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=70)
    if(k == 1):
        j += 1
        
```
