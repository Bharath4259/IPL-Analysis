```python
import pandas as pd
mdf = pd.read_csv("data/ipl_matches.csv")
print(mdf.head(5))
```

To Get values of perticular season (column)

```python
# To Get values of perticular season (column)
season_2008 = mdf.loc[mdf['season'] == 2008]
season_2008
```

Use .last_valid_index() to get last index

```python
# Use .last_valid_index() to get last index of a data frame
last_index = season_2008.last_valid_index()
print(last_index)
```

```python
#Data of last index
last_row = season_2008.iloc[[last_index]]
last_row
```

```python
# Get cell value with row number and column name
# row number = 57 
# column name = team1
last_row.loc[57].at['team1']  # use last_index instead od 57
```

```python
# Get teams details
teams = [last_row.loc[last_index].at['team1'], last_row.loc[last_index].at['team2']]
teams
```

```python
teams = [last_row[['team1']], last_row[['team2']]]
print(teams)
```

```python
winner = last_row[['winner']]
print(winner)
```

```python
matches = []
# print(mdf.head(2))
prev_index = 0
for year in range(2008, 2017):
    season = mdf.loc[mdf['season'] == year]
#     print(season)
    last_index = season.last_valid_index()   #.last_valid_index() will return last index
    last_row = mdf.iloc[[last_index]]                    #Data of last index
    teams = [year, last_row.loc[last_index].at['team1'], last_row.loc[last_index].at['team2'], last_row.loc[last_index].at['winner']]
    matches.append(teams)
    
# print(matches)
# Winners data of each IPL season
final_matches_df = pd.DataFrame(matches, columns=['year', 'team1', 'team2', 'winner_team'])
final_matches_df 
```
